```python
import pandas as pd
import numpy as np
data = pd.read_csv('E:\\QQ\\研究生课程小作业\\统计软件\\example1.csv')
data
```

```python
data['x6 '] = data['x2 ']/data['x1 ']
data
```

```python
data.loc['mean',:] =[np.mean(data[j]) for j in data]
data.loc['sd',:] =[np.std(data[j]) for j in data]
data.tail()
```

```python
import matplotlib.pyplot as plt
data1 = data.iloc[0:-2,:]
data1
```

```python
plt.rcParams['font.family']=['SimHei']# 显示中文

plt.figure(1)
plt.subplot(2,1,1)
plt.plot(data1['y '], label='民航客运')
plt.title('民航客运趋势')
plt.plot(data1['x3 '], label='铁路客运')
plt.title('铁路客运趋势')
plt.legend(bbox_to_anchor=(1,1))


plt.figure(2)
plt.subplot(2,1,2)
plt.plot(data1['y ']/data1['x3 '], 'g.-',label='客运比例')
plt.title('民航客运与铁路客运比例')
plt.legend(bbox_to_anchor=(1,1))
```
