```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split,GridSearchCV
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
# import jieba
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error,classification_report,roc_auc_score
from sklearn.ensemble import RandomForestClassifier

# 导入数据
data = pd.read_csv("E:\\QQ\\研究生课程小作业\\统计推断小作业\\bank-full.csv",sep=';')
```

```python
for i in data.columns:
	if type(data[i][0]) is str:
		print("unknown value count in "+i+":\t"+str(data[data[i]=='unknown']['y'].count()))
```

```python
# 数据清洗
    # 确定目标值和特征值
x = data.iloc[:,0:-1]
data.loc[data['y'] == 'no', i] = 0
data.loc[data['y'] == 'yes', i] = 1
y = data['y']
    # 处理缺失值（最一般方法，用列均值替换）
# for i in ['job','education','contact','poutcome']:
#     x[i].replace('unknown',x.)#####
# x[['job','education','contact','poutcome']].fillna(x['age'].mean(),inplace=True)
    # 划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=666,test_size=0.2)
```

```python
# 特征工程——（字典）特征提取
    ## 特征中出现类别符号，需要进行one-hot编码处理(DictVectorizer)   x.to_dict(orient="records") 将数组特征转换成字典数据
    # 先实例化一个转换器
transformer = DictVectorizer()
    # 由于dataframe无法提取特征，即需要转化为字典形式，才能继续提取特征
x_train = x_train.to_dict(orient='records')
x_test = x_test.to_dict(orient='records')

    # 调用fit_transform方法提取特征
x_train = transformer.fit_transform(x_train)
x_test = transformer.fit_transform(x_test)
# y_train = transformer.fit_transform(y_train)
# y_test = transformer.fit_transform(y_test)
```

```python
x_train
```

```python
# 模型训练（决策树）
    # 先实例化一个估计器
estimator = DecisionTreeClassifier(criterion='entropy',max_depth=5)     # 选择信息增益方法画树，最大树深限制为5
    # 调用fit进行训练
estimator.fit(x_train,y_train)
```

```python
# 模型预测
y_pre = estimator.predict(x_test)
print('预测值：\n',y_pre)
score = estimator.score(x_test,y_test)
print('预测准确率：\n',score)

    # 模型评价
mse = mean_squared_error(y_true=y_test,y_pred=y_pre)
print('均方误差：\n',mse)

evaluate = classification_report(y_true=y_test,y_pred=y_pre)
print('分类报告：精确率，召回率，F1—score\n',evaluate)
```

```python
    # 不平衡二分类的ROC曲线和AUC指标
    # 二分类模型的专门评价指标AUC
        # api：roc_auc_score(y_true=,y_score=)
        
y_true = np.where(y_test>0,1,0)      ### 注意y_true为样本的真实标签，y_score为预测为1的概率值
                                        #例如参数y_true = [0, 0, 1, 1, 1]为样本的真实标签
                                        #例如参数y_score = [0.1, 0.2, 0.7, 0.8, 0.9]为样本预测为1的概率值
auc_zhibiao = roc_auc_score(y_true=y_true,y_score=y_pre)
print('AUC指标为：\n',auc_zhibiao)
```
